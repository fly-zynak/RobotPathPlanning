#### Obstacles using polygons

```python
def obst(coord):
    coord.append(coord[0]) #repeat the first point to create a 'closed loop'
    xs, ys = zip(*coord) #create lists of x and y values
    plt.figure()
    plt.plot(xs,ys)    
obst([(1,1),(2,1),(2,2),(1,2)])
plt.show()
```

#### Defining Goal region

```python
def goal(coord):
    coord.append(coord[0]) #repeat the first point to create a 'closed loop'
    xs, ys = zip(*coord) #create lists of x and y values
    plt.figure()
    plt.plot(xs,ys)
    return xs,ys
    
goal([(6,6),(6,7),(7,7),(7,6)])
plt.show
```

```python
import random
import math
def RP(xf,yf):
    x = random.uniform(0,xf)
    y = random.uniform(0,yf)
    return (x,y)
def dist(x1,y1,x2,y2):
    d = math.sqrt((x1-x2)**2 + (y1-y2)**2)
    return d
```

## Trees (Will be used in final RRT code)

```python
class node():
    
   	#Generic tree node
    def __init__(self, name='root', ch=None, par = None):
        self.name = name
        self.ch = []
        if ch is not None:
            for child in ch:
                self.add_child(child)
        self.par = par

    def __repr__(self):
        return self.name
    def __str__(self):
        return self.name
    def add(self, node):
        #assert isinstance(node, Tree)
        self.ch.append(node)
        node.par = self
        
    def dist(x1,y1,x2,y2):
        d = math.sqrt((x1-x2)**2 + (y1-y2)**2)
        return d
        
    def nearest(self,xr,yr,md,pt):
        
        if md >= dist(self.name[0],self.name[1],xr,yr):
                md = dist(self.name[0],self.name[1],xr,yr)
                pt = self
                
        if len(self.ch)==0:
                return (md,pt)
        else:   
            for i in self.ch:
                a =  i.nearest(xr,yr,md,pt)
                if md >= a[0]:
                    md = a[0]
                    pt = a[1]
                else:
                    pass
            return (md,pt)   
    def get(self,level):
        if len(self.ch)==0:
            for i in range(level):
                print("-",end="")
            print(str(self.name)+"<-"+str(self.par.name))
            #print(str(self.name))
        else:
            for i in range(level):
                print("-",end="")
            if level !=0:
                print(str(self.name)+"<-"+str(self.par.name))
            else:
                print(str(self.name))
            for i in self.ch :
                i.get(level+1)
    def pl(self,level):
        #plt.figure(figsize=(15,10))
        ax = []
        ay = []
        if len(self.ch)==0:
            ax.append(self.par.name[0])
            ax.append(self.name[0])
            ay.append(self.par.name[1])
            ay.append(self.name[1])
            plt.plot(ax,ay,"r.-")
        else:
            if level ==0:
                pass
            else:
                ax.append(self.par.name[0])
                ax.append(self.name[0])
                ay.append(self.par.name[1])
                ay.append(self.name[1])
                plt.plot(ax,ay,"r.-")
            for i in self.ch :
                i.pl(level+1)
    def tb(self,n):
        ax = []
        ay = []
        ax.append(n.name[0])
        ay.append(n.name[1])
        while n.name != self.name:
            n = n.par
            ax.append(n.name[0])
            ay.append(n.name[1])
        plt.plot(ax,ay,"b.-")
```

# Rapidly expanding Random Trees Alogrithm

### RRT Algorithm to find a path

##### RRT Pseudo Code

    Qgoal #region that identifies success
    Counter = 0 #keeps track of iterations
    lim = n #number of iterations algorithm should run for
    G(V,E) #Graph containing edges and vertices, initialized as empty
    
    While counter < lim:

        Xnew  = RandomPosition()
        if IsInObstacle(Xnew) == True:
            continue
        Xnearest = Nearest(G(V,E),Xnew) //find nearest vertex
        Link = Chain(Xnew,Xnearest)
        G.append(Link)
        if Xnew in Qgoal:
            Return G
            
    Return G

### The final Program

```python
c = 0
xf,yf = map(int,input("Enter bounds").split())
xs, ys = map(int,input("Start:").split())
xe,ye = map(int,input("End:").split())
num = int(input("Number of obstacles?  "))
print("Input coordinates of vertices for each polygon.")
#plt.figure(figsize=(15,10))
p = []
for i in range(num):
    ver = int(input("No. of vertices: "))
    v = []
    for j in range(ver):
        x,y = map(float,input("x y: ").split())
        v.append((x,y))
    p.append(Polygon(v))


root = node((xs,ys))
last = root
print(root.name)
ac = 0
while c <= 5000:
    rp = RP(xf,yf)
    don = 0
    if c%120 == 0:
        if ac==0:
            ac = 1
            rp = (xe,ye)
            don = 1
        else:
            pass
    else:
        ac = 0
    #if Point(rp).within(p1):
    #    continue
    decider = 0
    for i in range(num):
        if Point(rp).intersects(p[i]):
            decider = 1
            break
    if decider == 1:
        continue
    
    nn = node(rp)
    #print(nn.name)
    g = root.nearest(rp[0],rp[1],dist(rp[0],rp[1],root.name[0],root.name[1]),root)
    if don == 0:
        if g[0]<=0.013*max(dist(xe,ye,xf,yf),dist(xf,yf,xs,ys),dist(xe,ye,xs,ys),100) or g[0]>=0.06*max(dist(xe,ye,xf,yf),dist(xf,yf,xs,ys),dist(xe,ye,xs,ys),100):
        #if g[0]>=0.06*max(dist(xe,ye,xf,yf),dist(xf,yf,xs,ys),dist(xe,ye,xs,ys),100):
            continue
    else:
        print("yea")
    y = g[1]
    if nn.name[0] == y.name[0] and y.name[1] == nn.name[1]:
        continue
    jl = LineString([nn.name,y.name])
    decider = 0
    for i in range(num):
        if jl.intersects(p[i]):
            decider = 1
            break
    if decider == 1:
        continue
    #print(y.name)
    y.add(nn)
    if nn.name[0] >= xe-1 and nn.name[0] <= xe +1:
        if nn.name[1] >= ye-1 and nn.name[1] <= ye +1:
            last = nn
            print("Success!")
            break
    if c%300 == 0:
        print(c)
    c+=1
#root.get(0)
plt.figure(figsize=(12,9))
print(xf,yf)
plt.plot([0-1,xf+1,xf+1,0-1,0-1],[0-1,0-1,yf+1,yf+1,0-1],"y-")
plt.plot([xe-1,xe+1,xe+1,xe-1,xe-1],[ye-1,ye-1,ye+1,ye+1,ye-1],"b")
for i in range(num):
    pol = p[i]
    #print(type(pol))
    x,y = pol.exterior.xy
    plt.plot(x,y,"g.-")

#plt.plot(x,y)
#plt.figure(figsize=(12,9))
root.pl(0)
root.tb(last)


"""
No. of vertices: 5
x y: 5 25
x y: 35 118
x y: 46 150
x y: 27 180
x y: 2 67
No. of vertices: 3
x y: 28 200
x y: 45 100
x y: 75 200
No. of vertices: 4
x y: 50 0
x y: 60 0
x y: 60 90
x y: 50 90
No. of vertices: 4
x y: 70 130
x y: 160 130
x y: 160 170
x y: 70 170
-------
No. of vertices: 4
x y: 130 20
x y: 140 20
x y: 140 300
x y: 130 300
No. of vertices: 4
x y: 100 0
x y: 110 0
x y: 110 280
x y: 100 280
No. of vertices: 4
x y: 5 120
x y: 100 120
x y: 100 150
x y: 5 150
"""        
```
