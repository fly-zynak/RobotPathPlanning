```python
import matplotlib.pyplot as plt
```

##### Line plots

```python
x = [1,3,4]
y= [8,2,4]
x2 = [10,20,50]
y2 = [4 , 6 ,9 ]
plt.plot(x,y,label = "First")
plt.plot(x2,y2,label="Second")
plt.xlabel("A")
plt.ylabel("B")
plt.title("Example")
plt.legend()
plt.show()
```

##### Scatter Plots

```python
x1 = [1,3,4]
y1= [8,2,4]
x2 = [10,20,50]
y2 = [4 , 6 ,9 ]
#plt.scatter(x1,y1,label= "1", )
plt.scatter(x2,y2, label = "2")
plt.plot(x1, y1,"r.-",label = "1")
plt.xlabel("A")
plt.ylabel("B")
plt.title("Example")
plt.legend()
plt.show()
```

#### Obstacles using polygons

```python
def obst(coord):
    coord.append(coord[0]) #repeat the first point to create a 'closed loop'
    xs, ys = zip(*coord) #create lists of x and y values
    plt.figure()
    plt.plot(xs,ys)    
obst([(1,1),(2,1),(2,2),(1,2)])
plt.show()
```
